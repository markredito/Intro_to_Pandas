# Grouping and Sorting

```python
spotifydata
```

```python
#groupwise analysis using by using "groupby"
spotifydata.groupby("track_name").popularity.count()
# the output should be an index of "track_name" with "popularity" next to it.
```

```python
#lets try another one
spotifydata.groupby("track_name").available_markets.count()
```

```python
#now lets group the values in the column "track_name" and show the "loudness" value
spotifydata.groupby("track_name").loudness.min()
```

You can think of each group we generate as being a slice of our DataFrame containing only data with values that match. This DataFrame is accessible to us directly using the `apply()` method, and we can then manipulate the data in any way we see fit. 

```python
spotifydata.groupby("tempo").apply(lambda df: df.loc[df.liveness.idxmax()])
```

```python
# agg() lets you run a bunch of different functions on your dataframe simultaneously.
spotifydata.groupby(["track_name"]).duration_min.agg([len, min, max])
# I was meaning to call the length, min and max of "duration" but since
# it only has one value, then min and max are the same. length is 1 because there's
# only one entry.
```

## Multi-indexes

```python
# my data might not be robust enough for this
# I'm trying to create a multi intex with track name as the first index, then
# popularity second, showing the length of "available_markets" in the next column
parental_advisory = spotifydata.groupby(["track_name", "popularity"]).available_markets.agg([len])
parental_advisory
```

```python
#this shows us that "parental_advisory" is a multi-index
mi = parental_advisory.index
type (mi)
```

Multi-indices have several methods for dealing with their tiered structure which are absent for single-level indices. They also require two levels of labels to retrieve a value. Dealing with multi-index output is a common "gotcha" for users new to pandas.

```python
#reset the index
parental_advisory.reset_index()
```

## Sorting

```python
# using the previous assignment "parental_advisory", we sort the data by "popularity"
parental_advisory.sort_values(by = "popularity")
```

`sort_values()` defaults to an ascending sort, where the lowest values go first. However, most of the time we want a descending sort, where the higher numbers go first. Here's how to do it:

```python
#same as above but this time in descending order
parental_advisory.sort_values(by = "popularity", ascending = False)
```

```python
#sorting by index value. So you see here that our index is "track_name",
#sorted alphabetically starting from symbols and numbers first.
parental_advisory.sort_index()
```

```python
#we can sort more than one column at a time:
parental_advisory.sort_values(by=["track_name", "popularity"])
```
