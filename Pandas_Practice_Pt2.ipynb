```python
#lets import pandas and load a new data set
import os
import pandas as pd
```

```python
#load the dataset
mh = pd.read_csv("E:\CODE\Pandas\smmh.csv")
mh
```

```python
# we can use `dtype` to grab the type of a specific column.
# here we pull up the `dtype` of "gender"
mh.Timestamp.dtype
```
