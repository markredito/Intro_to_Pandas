# Renaming
- `rename()` allows you to change index names and/or column names.

```python
#lets rename the columns in our dataset
renamed = mh.rename(columns={"1. What is your age?":"age"})
renamed
```

```python
#lets rename a few more columns
renamed = mh.rename(columns={"1. What is your age?":"age",
                            "2. Gender":"gender", 
                            "3. Relationship Status":"rel_status",
                            "4. Occupation Status":"occupation"})
renamed
```

```python
# lets rename the index
renamed_index = renamed.rename(index={0: "first_entry", 1: "second_entry"})
renamed_index
```

You'll probably rename columns very often, but rename index values very rarely. For that, `set_index()` is usually more convenient.

Both the row index and the column index can have their own `name` attribute. The complimentary `rename_axis()` method may be used to change these names.

```python
# lets rename the index name of rows to "entries" and the columns to "fields"
renamed_idx2 = renamed_index.rename_axis("entries", axis = "rows").rename_axis("fields", axis ="columns")
renamed_idx2
```

# Combining
Three methods:
- `concat()`- given a list of elements, this function will smush those elements together along an axis
- `join()`
- `merge()`

```python
#load new datasets
brands = pd.read_csv("brands.csv")
similar = pd.read_csv("similarweb.csv")

#concatenate both dataframes
concat_brands = pd.concat([brands, similar])

concat_brands
```

```python
#using join()

a = brands.set_index(["name", "site"])
b = similar.set_index(["name", "site"])

a.join(b, lsuffix = "aaa", rsuffix = "bbb")

a
```
